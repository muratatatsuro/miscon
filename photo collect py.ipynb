```python
#photo collwctor py
import os
import bs4
import time
import random
from itertools import chain
import urllib.request as req

base_url='https://misscolle.com'
"""base_url
   htmlの取得
   bs4.beautifulsoup(html,'html.parser')
   columns=soup.find_all('ul',class_='columns)
   atags=map(lambda x:x.find_all('a'),columns)
   
   for tag in chain.from_iterable(atags):
   #属性を取得
   path=tag.get('href')
   属性がhttpじゃないならばパスを作成
   if path.startswith('http'):
   path='{}{}'.format(base_url,path)
   if path[-1]=='/':
   path=path[:-1]
   
   f.write('{}\n'.format(path))
   page_urls.txtの出来上がり
   """

"""page_urls.txtを作成するための関数"""
def fetch_page_urls():
    html=req.urlopen('{}/versions'.format(base_url))
    #beautifulsoup4を用いる
    soup=bs4.BeautifulSoup(html,'html.parser')
    
    #特定のタグを抽出
    columns=soup.find_all('ul',class_='columns')
    atags=map(lambda x:x.find_all('a'),columns)
    
    #画像のあるページのURLが書かれているテキストを作成
    with open('page_urls.txt','w') as f:
        #リストの中の文字を全て順序良く取り出す
        for st in chain.from_iterable(atags):
            #取得したタグの属性を取得
            path=st.get('href')
            #urlの最初の文字がhttpかどうか
            if not path.startswith('http'):
                path='{}{}'.format(base_url,path)
            #pathの最後一文字が'/'ならば   
            if path[-1]=='/':
                path=path[:-1]
                
            f.write('{}\n'.format(path))
            
#写真を集めていく
def fetch_photo():
    #page_urls.txtを読み込む
    with open('page_urls.txt','r') as f:
        for url in f:
            #写真を保存しておくdirectoryを作成
            
            #directory path
            dirpath=r'D:/photos/{}'.format(url.strip().split('/')[-1])
            os.mkdir(dirpath)
            
            html=req.urlopen('{}/photo'.format(url.strip()))
            soup=bs4.BeautifulSoup(html,'html.parser')
            
            photos=soup.find_all('li',class_='photo')
            paths=map(lambda path:path.find('a').get('href'),photos)
            
            for path in paths:
                #?
                filename='_'.join(path.split('?')[0].split('/')[-2:])
                #filepathを作成
                filepath='{}/{}'.format(dirpath,filename)
                
                #download image
                #ダウンロードしたデータをファイルに移送
                req.urlretrieve('{}{}'.format(base_url,path),filepath)
                #add random waiting time
                time.sleep(4+random.randint(0,2))
                
if __name__=='__main__':
    fetch_page_urls()
    fetch_photo()
            
        
    
```
