```python
#faec detect
import matplotlib.pyplot as plt
import cv2
import os

"""srcpathは写真フォルダが存在しているdirectoryへのパス
   顔を認識する関数"""
def face_detect(srcpath,dstpath,filename):
    cascade_file='haarcascade_frontalface_alt2.xml'
    cascade=cv2.CascadeClassifier(cascade_file)
    
    #画像の取り込み
    img=cv2.imread('{}/{}'.format(srcpath,filename))
    img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    faces=cascade.detectMultiScale(img_gray)
    
    if len(faces)==1:
        (x,y,w,h)=faces[0]
        images=img[y:y+h,x:x+w]
        #トリミング
        image=cv2.resize(images,(100,100))
        #保存
        cv2.imwrite('{}/{}'.format(dstpath,filename),image)
    

def main():
    """rootはdirectoryまでのパスが入っているリスト
    　　dirsはdirectoryのパスが入っているリスト
      filesはfilenameが入っているリスト"""
    
    
    for root,dirs,files in os.walk(r'D:/photos1'):
        srcpath=root
        if len(dirs):
            continue
        #画像を保存するdirectoryはphotosからfacesに変化   
        dstpath=srcpath.replace(r'D:/photos1',r'D:/faces')
        print(dstpath)
        if not os.path.isdir(dstpath):
            os.mkdir(dstpath)
        
        for filename in files:
            #エラー処理
            if filename.startswith('.'):
                continue
            try:
                face_detect(srcpath,dstpath,filename)
            except:
                continue
        
if __name__=='__main__':
    main()
    
```

```python
import cv2
import os
import matplotlib.pyplot as plt
%matplotlib inline
cascade_file='haarcascade_frontalface_alt2.xml'
cascade=cv2.CascadeClassifier(cascade_file)
    
#画像の取り込み
img=cv2.imread('3_1.jpg')
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
faces=cascade.detectMultiScale(img_gray)
    
if len(faces)==1:
    (x,y,w,h)=faces[0]
    images=img[y:y+h,x:x+w]
    #トリミング
    image=cv2.resize(images,(100,100))
    #保存
    cv2.imwrite('3_2.jpg',image)
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.show()
```
